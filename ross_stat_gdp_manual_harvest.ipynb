```python
import pandas as pd
```

```python
url = 'https://rosstat.gov.ru/storage/mediabank/152_09-09-2022.html'
tables = pd.read_html(url)
```

```python
data= tables[0].iloc[1:]
data
```

```python
#take only dta for 2022 and skip 'half year data', keeping quarters only
data_2022 = data.iloc[-3:-1, :2].dropna() 
data_2022 .columns=['Quarter', 'gdp_growth']
```

```python
#changes made for match WOrld Bank format
data_2022['gdp_growth_WBstyle']=(data_2022['gdp_growth'].astype(int)/10) #reintroduce decimal point
```

```python
data_2022['gdp_growth_WBstyle'].astype(float)
```

```python
data_2022['gdp_growth_WBstyle'] =data_2022['gdp_growth_WBstyle']-100
```

```python
data_2022
```

Q1 refers to Quarter 1 (Jan to Mar), Q2 refers to Quarter 2 (Apr to June), Q3 refers to Quarter 3 (July to Sept) and Q4 refers to Quarter 4 (Oct to Dec).

```python
quarters  = ['2022-01-1','2022-06-01']


data_2022['year']= quarters
```
