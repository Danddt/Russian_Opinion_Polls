```python
# LIST OF CONFLICTS UKRAINE RUSSIA

import pandas as pd
import numpy as np

link = "https://en.wikipedia.org/wiki/List_of_wars_involving_Ukraine"
tables = pd.read_html(link)

#data= tables[2].iloc[1:]
```

```python
listo = []
for item in range(1, 8):
  listo.append(tables[item])
```

```python
complete = pd.concat(listo)
```

```python
clean =complete.reset_index().drop(['index'], axis=1)
```

```python

# if column contains : Russian, Russia, Soviet
mylist = ['Russian', 'Russia', 'Soviet']
pattern = '|'.join(mylist)

clean_short =clean[clean['Combatant 2'].str.contains(pattern)]
```

```python
clean_short['Date'] = clean_short['Date'].str.split(' ', expand=True).drop([1,2], axis=1)
clean_short
```

```python
clean_short = clean_short.sort_values(by='Date')
clean_short_short = clean_short[['Date', 'Conflict']]
```

```python
clean_short_short.reset_index().drop(columns=['index'])
```

```python
clean_short_short.to_csv(str(new_dir)+ '/conflict_list.csv', index=False)
```
