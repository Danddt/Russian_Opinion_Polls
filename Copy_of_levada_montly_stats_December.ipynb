```python
import pandas as pd
import numpy as np
```

```python
moscow1 = pd.read_csv('/content/drive/MyDrive/RUSSdata/MoscowTimes/internet_fake_articles.csv')
moscow2 = pd.read_csv('/content/drive/MyDrive/RUSSdata/MoscowTimes/internet_search_articles.csv')
censorship = pd.read_csv('/content/drive/MyDrive/movie_russia/December_2022/censorshipUptoMarch.csv')
```

```python
key_dumi =['block', 'internet', ]

key_dumi_regex = "|".join(key_dumi)
```

```python
moscow2['date'] = pd.to_datetime(moscow2['date'])
```

```python
articles =moscow2.groupby(moscow2.date.dt.year)['topic'].count().reset_index()
```

```python
# as not filtered halving the possible score 
articles['score:topic*.5'] = articles['topic']*.5
```

```python
articles.set_index('date', inplace=True)
```

```python
import numpy as np
key_words =['filtering', 'implementation', 'blacklist', 'block', 'law', 'restricted', 'ban', 'isolate', 
            'fine', 'harrasment', 'criminal','throttling', 'step down', 'resign', 'killings', 'harassed', 'suppress', 'detained', 'arrested']
key_words2 = ['jail', 'sentenced', 'prison']

key_words_regex = "|".join(key_words)
key_words2_regex = "|".join(key_words2)

censorship['arbitrary_score'] = np.add((np.where(censorship['event'].str.contains(key_words_regex, regex=True), 1, 0.1)), 
       (np.where(censorship['event'].str.contains(key_words2_regex, regex=True), 2, 0.1)))
```

```python
censorship['arbitrary_score'] =censorship['arbitrary_score'].astype(str).astype(float)
censorship['event_occurances'] =censorship['event_occurances'].astype(str).astype(float)
```

```python
censorship['score*events'] = censorship['event_occurances']*censorship['arbitrary_score']
```

```python
censorship['datetime']= pd.to_datetime(censorship['datetime'])
```

```python
grouped = censorship.groupby(censorship['datetime'].dt.year)['score*events'].sum().reset_index()
grouped.set_index('datetime', inplace=True)
```

```python
group2 = grouped.join(articles).fillna(0)
group2 = group2[['score*events', 'score:topic*.5']]
group2['total_score_year'] = group2['score*events']+group2['score:topic*.5']
group2.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime
import seaborn as sns

fig, ax = plt.subplots( figsize=(14, 7), constrained_layout=True)


sns.barplot(x = group2.index, y = group2['total_score_year'])

plt.ylabel('Events count',fontsize=14)
plt.xlabel('Years',fontsize=14)

plt.title('Censorship events: grouped by Year', fontsize = 16)
plt.grid()

plt.savefig('/content/drive/MyDrive/movie_russia/December_2022/censorhsip_bar.jpg')
```

```python
group2['test'] = (np.log(group2['total_score_year'].cumsum()))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime
import seaborn as sns

fig, ax = plt.subplots( figsize=(12, 7), constrained_layout=True)


plt.plot(group2.index, group2['test'].rolling(2).mean(), linewidth = 3, color = 'red')
plt.title('Censorship curve : Moving Average (on logarithmic values)',fontsize=18)
plt.ylabel('Moving Average (log values)',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.grid()
plt.savefig('/content/drive/MyDrive/movie_russia/December_2022/censorhsip_LOG_line.jpg')
```

```python
import os
df_read= []

diro= '/content/drive/MyDrive/RUSSdata/Levada/information_trust/2023-01-04/'

listo1 = []
for filename in os.listdir(diro):
  if filename.startswith('TRANS'):
    if filename.endswith('_0.csv'):
      listo1.append(filename)
    if filename.endswith('_2.csv'):
      listo1.append(filename)
    if filename.endswith('_3.csv'):
      listo1.append(filename)
    if filename.endswith('_4.csv'):
      listo1.append(filename)

print(listo1)
```

```python
def clean_up1 (data):
  data.columns = data.iloc[0]
  data1 = data.iloc[1:, :]
  data1 = data1.apply(pd.to_numeric,errors='ignore')
  return data1
```

```python
df_using = []
for df in listo1:
  df_read= pd.read_csv(diro+df)
  df_read= clean_up1 (df_read)
  df_using.append(df_read)

```

```python
fake_law =  df_using[0]
fake_law_aims =  df_using[1]
noticed_changes_in_media =  df_using[2]
who_u_trust = df_using[3]
```

```python
fake_law_support = fake_law.iloc[:2,2:-1]
#fake_law_support.loc['row_total', :] = fake_law_support.sum().values
fake_law_support_age =fake_law_support.T.reset_index()
```

```python
fake_law_support_age.columns=['age group','I fully support','Rather support' ]
```

```python
fake_law_support_age
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns

fig, ax = plt.subplots( 1,2, figsize=(22, 12), constrained_layout=True)
#define data
data = fake_law.Total
labels = fake_law['in']
plt.ylim(0,100)
#define Seaborn color palette to use
colors = sns.color_palette('pastel')

#create pie chart
ax[0].pie(data, labels = labels, colors = colors, autopct='%.0f%%',textprops={'fontsize': 14})
ax[0].set_title('Support Overall', fontsize=16)


g = fake_law_support_age.plot(kind='bar',stacked = True, color=colors, ax=ax[1])

for c in g.containers:
    g.bar_label(c,label_type='center')

# labels for x & y axis
ax[1].set_xticklabels([])
plt.xlabel('Score')
plt.ylabel('Age Group')

ax[1].set_title('Support by Age (survey values, stacked)', fontsize=16)
ax[1].legend(loc='upper left', fontsize =14)
plt.grid(axis='y')
fig.suptitle('DO YOU SUPPORT OR OPPOSE THE LAW ON CRIMINAL LIABILITY FOR FALSE INFORMATION \n ABOUT THE ACTIONS OF THE RUSSIAN MILITARY \n IN THE FRAMEWORK OF A SPECIAL OPERATION AND \nCALLS FOR ANTI-RUSSIAN SANCTIONS? (in %)\n\n', fontsize=16)

plt.savefig('/content/drive/MyDrive/movie_russia/December_2022/fake_law_support.jpg')
```
